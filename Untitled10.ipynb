# Model Training

```python
X = df.drop(['Survived'],axis=1)
Y = df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model1 = SVC(kernel = 'rbf')
model2 = RandomForestClassifier()
model3 = LogisticRegression()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
ypred1 = model1.predict(x_test)
ypred2 = model2.predict(x_test)
ypred3 = model3.predict(x_test)
```

```python
cm1 = confusion_matrix(y_test,ypred1)
cm2 = confusion_matrix(y_test,ypred2)
cm3 = confusion_matrix(y_test,ypred3)
```

```python
print(f"Model performance of Support vector machine:\n\n\n{classification_report(y_test,ypred1)}\n\n\n{sns.heatmap(cm1,annot = True)}")
```

#  Titanic Dataset Analysis and Model Performance Report

## 1. Data Analysis
The analysis began with loading the **Titanic dataset**.  
An initial inspection revealed missing values in several columns  notably **Age**, **Cabin**, and **Embarked**.

### Handling Missing Data
- The missing values in **Age** were imputed with the **mean age**.
- Rows with missing **Cabin** and **Embarked** values were **dropped** to maintain data quality.

### Feature Engineering
A new feature, **`family_size`**, was created by combining the **SibSp** and **Parch** columns:
\[
\text{family\_size} = \text{SibSp} + \text{Parch}
\]

### Data Cleaning
Irrelevant or non-informative columns were removed to simplify the model:
- **Dropped Columns:** `PassengerId`, `Name`, `Ticket`, and `Fare`

### Encoding Categorical Variables
Categorical features  **Sex**, **Cabin**, and **Embarked**  were transformed into numerical form using **Label Encoding** to make them suitable for machine learning algorithms.

---

## 2. Exploratory Data Analysis (EDA)
Exploratory analysis focused on understanding the relationships between key features and the **Survived** target variable.

- **Scatter Plots** were generated to visualize relationships between numerical features and survival outcomes.
- **Correlation analysis** was performed to identify features with the highest impact on survival.

The analysis showed that the feature **Sex** had the **strongest correlation** with the target variable **Survived**, indicating that female passengers had a significantly higher survival rate.

---

## 3. Model Training
Three classification models were trained to predict survival outcomes:

1. **Support Vector Classifier (SVC)**  with a radial basis function (RBF) kernel  
2. **Random Forest Classifier**  
3. **Logistic Regression**

The dataset was divided into:
- **Training Set:** 80% of the data  
- **Test Set:** 20% of the data  

A fixed random state of **42** ensured reproducibility.  
All models were trained on the processed training data.

---

## 4. Model Evaluation
Each models performance was evaluated using:
- **Classification Report** (Precision, Recall, F1-Score, and Support)
- **Confusion Matrix**

### 4.1 Support Vector Classifier (SVC)
- The **SVC** model showed **poor performance**, with a precision of **0.00** for class `0` (*not survived*).
- It predicted nearly all instances as **survived**, resulting in an imbalanced confusion matrix.
- The recall for class `1` (*survived*) was high, but overall accuracy and balance were low.

### 4.2 Random Forest Classifier
- The **Random Forest** model demonstrated **better performance** than SVC.
- It achieved **balanced precision and recall** for both survival classes.
- The confusion matrix indicated a fair distribution of correct predictions across both categories.

### 4.3 Logistic Regression
- **Logistic Regression** achieved the **best overall performance** among the three models.
- It delivered **high precision, recall, and F1-score** for both survived and non-survived classes.
- The confusion matrix showed a good balance between true positives and true negatives, indicating effective model generalization.

---

## 5. Summary and Insights
- **Best Model:** Logistic Regression  
- **Key Findings:**
  - The feature **Sex** was the most influential in determining survival.
  - Proper handling of missing values and encoding categorical data significantly improved model accuracy.
  - **Logistic Regression** outperformed both **SVC** and **Random Forest** in this feature setup.

### Final Conclusion
Based on the evaluation metrics and confusion matrices, **Logistic Regression** is identified as the **best-performing model** for this dataset, providing a strong balance between interpretability and predictive accuracy.
